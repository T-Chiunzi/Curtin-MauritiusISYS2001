## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### üí° Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
You said
Can you make a program in simple Python language that allows users to input their monthly income, allows users to input up to 3 expenses, calculates the remaining budget and outputs a warning if the budget is negative?
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# 1. Get the Monthly Income
income = float(input("Enter your monthly income: "))

# 2. Get 3 Expenses
print("Please enter your 3 main expenses:")
expense1 = float(input("Expense 1: "))
expense2 = float(input("Expense 2: "))
expense3 = float(input("Expense 3: "))

# 3. Calculate the Total and Remaining Budget
total_expenses = expense1 + expense2 + expense3
remaining_budget = income - total_expenses

# 4. Output the Results
print("\n--- Monthly Budget Summary ---")
print("Total Income: $" + str(income))
print("Total Expenses: $" + str(total_expenses))
print("Remaining Budget: $" + str(remaining_budget))

# 5. Warning Logic
if remaining_budget < 0:
    print("WARNING: You are over budget! Your expenses exceed your income.")
else:
    print("Great job! You are within your budget.")
```
